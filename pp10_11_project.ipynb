```python
def insert_data(df, new_df):
    df = pd.concat([df, new_df], ignore_index=True)
    return df
```

Search and Filter Features

```python
def search_book_title():
  title = input("Enter the book title: ")
  return title

def filter_df(df, column_name, search_key):
  searched_df = df[df[column_name] == search_key]
  if searched_df.empty:
    return f"No data found with search key: {search_key}"
  else:
    return searched_df

df.head()

search_by_title = input("Enter the title of the book you want: ")

```

```python
filtered_df = filter_df(df, 'Book Title', search_by_title)
print(filtered_df)
```
